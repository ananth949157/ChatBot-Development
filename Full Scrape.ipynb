```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin

# Define a function to process and filter links
def process_links(url, filter_prefix):
    # Send an HTTP GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all the anchor tags (<a>) in the HTML
        anchor_tags = soup.find_all('a')

        # Initialize a set to store unique valid hrefs
        valid_hrefs = set()

        # Iterate through the anchor tags and process the href attributes
        for anchor in anchor_tags:
            href = anchor.get('href')
            if href:
                # Concatenate the href with the base URL
                full_url = urljoin(url, href)
                
                # Check if the full URL starts with the specified prefix
                if full_url.startswith(filter_prefix):
                    valid_hrefs.add(full_url)

        # Convert the set back to a list and return the unique valid hrefs
        return list(valid_hrefs)
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return []

# Define the URL of the page you want to start from
starting_url = "https://www.usf.edu/business/"

# Process and filter the links starting from the starting URL
filtered_links = process_links(starting_url, "https://www.usf")

# Function to recursively process and collect links
def collect_links_recursive(urls, filter_prefix):
    collected_links = []

    for url in urls:
        # Process and filter links for the current URL
        links = process_links(url, filter_prefix)
        
        # Add the links to the collected_links list
        collected_links.extend(links)

    return collected_links

# Collect links recursively starting from the initial URL and filtered links
all_links = collect_links_recursive([starting_url] + filtered_links, "https://www.usf")

# Print all the collected links
for link in all_links:
    print(link)
```

```python
# Iterate through the links
for index, link in enumerate(all_links):
    try:
        # Send an HTTP GET request to the URL
        response = requests.get(link)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content of the page
            soup = BeautifulSoup(response.text, 'html.parser')

            # Find the div with class "mainContent_well u-flexItem--largeExtra"
            div_content = soup.find('div', class_='mainContent_well u-flexItem--largeExtra')

            # Extract the text inside the div
            if div_content:
                page_text = div_content.get_text()

                # Define the file name based on the index
                file_name = f'page{index + 1}.txt'

                # Save the text to a text file
                with open(os.path.join('output_files', file_name), 'w', encoding='utf-8') as file:
                    file.write(page_text)

                print(f'Saved content from {link} to {file_name}')
            else:
                print(f'Could not find the specified div in {link}')

        else:
            print(f'Failed to retrieve content from {link}. Status code: {response.status_code}')

    except Exception as e:
        print(f'An error occurred while processing {link}: {str(e)}')
```
